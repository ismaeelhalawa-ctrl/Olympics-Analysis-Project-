```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/athlete_events.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull()
```

```python
print(f'THE NUMBER OF MALES IS : ' , df['Sex'].loc[df['Sex'] == 'M'].count())
print(f'THE NUMBER OF FEMALES IS : ' , df['Sex'].loc[df['Sex'] == 'F'].count())
print(f'THE NUMBER OF MALES AND FEMALES IS : ' , df['Sex'].count())
```

```python
df['Sex'].groupby(df['Year']).count()
```

```python
print('THE NUMBER OF CITIES IS : ' , df['City'].nunique())
print('AND THE CITIES ARE : ' , df['City'].unique())
```

```python
df['Name'].groupby(df['Year']).count()
```

```python
df[['Name' , 'ID']].value_counts().head(1)
```

```python
N_Gold =   df['Year'].loc[df['Medal'] == 'Gold'].value_counts().sort_index()
N_Silver =   df['Year'].loc[df['Medal'] == 'Silver'].value_counts().sort_index()
N_Bronze =   df['Year'].loc[df['Medal'] == 'Bronze'].value_counts().sort_index()
ct = pd.DataFrame([N_Gold, N_Silver, N_Bronze]).T
ct.columns = ['N_Gold', 'N_Silver', 'N_Bronze']
ct
```

```python
plt.plot(ct , marker = 'o' , lw = 1)
plt.title('Number Of Medals Each Year')
plt.xlabel('Year')
plt.ylabel('Number Of Medals')
plt.grid()
```

```python
st = df['City'].loc[df['Medal'] == 'Gold'].value_counts()
st
```

```python
plt.plot(st , marker = 'o' , lw = 1)
plt.title('Number Of Medals Each City')
plt.xlabel('Year')
plt.ylabel('Number Of Medals')
plt.grid()
```

```python
df.groupby('Year').apply(lambda x: x.isna().sum().sum())
```

```python
plt.plot(df.groupby('Year').apply(lambda x: x.isna().sum().sum()) , color = 'b' , marker = 'o' , mfc = 'y')
plt.xlabel('Year')
plt.ylabel('Number Of Missing Data')
plt.title('Number Of Missing Data Each Year')
plt.grid()
```
